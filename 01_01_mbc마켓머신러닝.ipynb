머신러닝 알고리즘은 지도학습과 비지도학습으로 나눠진다
1. 지도학습 : 훈련하기 위한 데이터와 정답이 필요
2. 비지도학습 : 정답이 없는 학습 자료가 필요하다.

지도학습에서 데이터와 정답을 코드로 입력해서 전처리 해야함
이떄 데이터르르 input이라는 변수에 넣고 정답을 target이라는 변수에 넣어 진행 하겠다

이 input과 target을 합쳐서 훈련데이터 (training data)라고 한다

입력으로 사용된 길이와 무게 -> 특성(feature)라고 한다

대부분 수치자료를 이용해서 판단하고 학습하며 분석할 수 있는 기법을 알아보자.

```python
# mbc 마켓에서는 살아있는 생선을 최초로 판매하려고 한다.
# 고객이 온라인 주문을 하면 가장 빠른 물류 센터에서 신선한 생선을 곧 바로 배송
# 여기서 문제 발생 !! ->  생선 포장 직원이 생선의 종류를 알지 못해?(넌 퇴사해라)
# IP 카메라를 이용해서 수족관을 실시간 촬영을 하고 이것을 객체 추적 기술로 활용;

# 현재는 AI 기술이 배우는 단계라 사이즈를 이용해서 AI한테 학습을 시켜보겠다;

# 일단 도미와 빙어부터 시작하겠음.(2개의 객체를 찾도록 훈련 -> 머신러닝)
# 전처리 데이터 필요

# 도미 생선의 길이 : bream_length 도미용 길이 리스트
bream_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0]

# 도미 생선의 무게 : bream_weigh 도미용 무게 리스트
bream_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0]
```

```python
# 특성 : 데이터의 특징(수치자료, 이미지 자료, 배열자료 등등...)
# 첫번째 도미의 특성 : 길이(25.4CM), 무게는(242.0g)
# 두번째 도미의 특성 : 길이(26.3CM), 무게는(290.0g)

# 두 특성을 숫자로 보는 것 보다 그래프로 표현하면 데이터를 잘 이해할 수 있다.
# 길이는 : X축/ 무게는 : Y축 -> 산점도(두 좌표의 관계)

# 파이썬의 외부라이브러리를 이용해서 차트 표시용 코드
import matplotlib.pyplot as plt
# 과학 계산용으로 그래프를 그리는 대표적인 패키지 : matplotlib
# https://wikidocs.net/92071
plt.scatter(bream_length, bream_weight)
# 가로길이 배열과 무게 배열을 차트로 만듬;
plt.xlabel('length') # X축 라벨
plt.ylabel('weight') # Y축 라벨
plt.show()

```

```python
# 보고서를 작성한다.
# 도미 35마리를 2차원 그래프에 점으로 나타내 보았다.
# 결론 : 생선의 길이가 갈수록 무게가 많이 나간다. 선현적이라고 함!
```

```python
# 빙어 데이터를 14개 준비
# 빙어 생선의 길이 : bream_length 빙어용 길이 리스트
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

# 빙어 생선의 무게 : bream_weigh 빙어용 무게 리스트
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()

```

```python
# 빙어 데이터를 14개 준비
# 빙어 생선의 길이 : bream_length 빙어용 길이 리스트
smelt_length = [9.8, 10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]

# 빙어 생선의 무게 : bream_weigh 빙어용 무게 리스트
smelt_weight = [6.7, 7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

plt.scatter(smelt_length, smelt_weight)
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
# 지금까지 리스트에 나누어져 있던 배열을 AI 학습용으로 재배치
length = bream_length + smelt_length # 도미의 길이 + 빙어의 길이
weight = bream_weight + smelt_weight # 도미의 무게 + 빙어의 무게

print(length)  # 물고기의 길이!
print(weight)  # 물고기의 무게!
```

```python
# zip()나열된 리스트 각각에서 하나씩 원소를 꺼내어 반환
fish_date = [[l,w]for l,w in zip(length, weight)]
# length, length를 2차원 리스트로 만듦
print(fish_date) # 2차원 리스트 출력
```

```python
# 머신러닝은 정답을 target이라는 값으로 알려줘야 한다.
# target 0(빙어), 1(도미)을 처리하려함

# 생선49개에 대한 정답 리스트를 만들어야함
# 머신러닝에서는 정답 데이터를 2진 분류로 테스트 해봄(1 : 도미, 0 : 빙어)

fish_target = [1] * 35 + [0] * 14
print(fish_target)
```

```python
# 지금까지 전처리 데이터를 만들었다면  AI에 기본적으로 기초가 되는 사이킷 런을 활용하여 트레이닝 시켜보자.
# https://wikidocs.net/26300 (코랩은 AI 용 외부라이브러리가 내장되어있음)
# 로컬 idle 에 실행하려면 pip install 등으로 설치 후 동작

from sklearn.neighbors import KNeighborsClassifier # 사이킷런 패키지안에 이웃패키지

kn = KNeighborsClassifier() # import된 객체를 생성하여 변수에 연결
```

```python
# 객체를 호출하였으면 전처리 데이터를 삽입한다.

# fish_data, fish_target을 전달하여 도미를 찾기 위한 기준을 학습 시킨다
# (트레이닝 데이터 : 훈련정답데이터)
kn.fit(fish_date, fish_target)
```

```python
# kn 훈련이 잘되었는지 평가 score()
kn.score(fish_date, fish_target)
# 훈련시 전처리 데이터가 얼마되지 않아 100% 나옴
```

```python
# k-최근접 이웃 알고르짐 파악해보기
# 어떤 데이터에 대한 답을 구할 때 주위의 다른 데이터를 보고 다수를 차지하는것을 정답으로 출력

plt.scatter(bream_length, bream_weight)
plt.scatter(smelt_length, smelt_weight)
plt.scatter(30, 600, marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
kn.predict([[30,600]])
```

```python
kn.predict([[10,10]])
```

```python
# 새로운 데이터에 대해 예측할때 가장 가까운 직선거리에 어떤 데이터가 있는지 살핌
# 단점, 데이터가 아주 많은 경우 사용하기 어려움
# (데이터가 크기에 메모리가 많이 필요하며 직선거리 계산 시간도 오래걸림)
# 객체를 메모리에 만들고 활용한다.



print(kn._fit_X)
```

```python
print(kn._fit_method)
```

```python
print(kn._y)
```

```python
# 결론 k-최근접 이웃 알고리즘은 훈련결과 파일이 존재하지 않는다.
# 새로운 데이터가 들어오면 가장 가까운 데이터를 메모리에서 참고한다
kn.predict([[5,15]])
```

```python
# 객체 생성시 n_neighbors= ??? 참고할 이웃 데이터 기본값은 5
# 이웃객체 4개 찾아서 결론을 낸다.

# 만약 5개가 아닌 다른 값으로 변경 할 수 있음

kn49 = KNeighborsClassifier(n_neighbors=49) # 49개를 참조
# 현재 원본타겟, 원본데이터가 49인데 가장가까운 49개 데이터 참조함

kn49.fit(fish_date, fish_target)
kn49.score(fish_date, fish_target)
```
