```python
# k-최근점 이웃 알고리즘의 한계에 대안 -> 선형회귀
# 학습한 k-최근접 이웃 회귀는 최대값에 대한 예측이 어렵다..

# 선형회귀는 비교적 간단하고 성능이 뛰어남
# 선형이이라는 말은 직선을 학습하는 알고리즘의
# 빨간선이 가로로 되어 있으면 농어의 무게를 1개로 인식함
# 빨간선이 위에서 시작해서 아래로 떨어지면 반대로 예측
# 빨간선이 아래에서 시작해서 위로 올라가면 정상적인 차트로 구현
```

```python
# 인어의 샘플자료 56개
import numpy as np

mermaid_length = np.array([8.4, 13.7, 15.0, 16.2, 17.4, 18.0, 18.7, 19.0, 19.6, 20.0, 21.0,
       21.0, 21.0, 21.3, 22.0, 22.0, 22.0, 22.0, 22.0, 22.5, 22.5, 22.7,
       23.0, 23.5, 24.0, 24.0, 24.6, 25.0, 25.6, 26.5, 27.3, 27.5, 27.5,
       27.5, 28.0, 28.7, 30.0, 32.8, 34.5, 35.0, 36.5, 36.0, 37.0, 37.0,
       39.0, 39.0, 39.0, 40.0, 40.0, 40.0, 40.0, 42.0, 43.0, 43.0, 43.5,
       44.0]) # 인어의 길이 배열리스트
mermaid_weight = np.array([5.9, 32.0, 40.0, 51.5, 70.0, 100.0, 78.0, 80.0, 85.0, 85.0, 110.0,
       115.0, 125.0, 130.0, 120.0, 120.0, 130.0, 135.0, 110.0, 130.0,
       150.0, 145.0, 150.0, 170.0, 225.0, 145.0, 188.0, 180.0, 197.0,
       218.0, 300.0, 260.0, 265.0, 250.0, 250.0, 300.0, 320.0, 514.0,
       556.0, 840.0, 685.0, 700.0, 700.0, 690.0, 900.0, 650.0, 820.0,
       850.0, 900.0, 1015.0, 820.0, 1100.0, 1000.0, 1100.0, 1000.0,
       1000.0]) # 인어의 무게 배열리스트
```

```python
# 사이킷 런을 이용해서 훈련세트와 테스트 세트를 나눔
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(mermaid_length, mermaid_weight, random_state=42)

# 2차원 배열로 만듬 -1을 넣으면 원소 개수로 모두 채움  ,1은 두번째 크기를 1로 만듦(특성 부여)
train_input = train_input.reshape(-1, 1)
test_input = test_input.reshape(-1, 1)

print(train_input) # 인어 키(훈련용)
```

```python
print(test_input)# 인어 키(테스트)
```

```python
print(train_target) # 인어 몸무게(훈련 정답용)
```

```python
print(test_target)# 인어의 몸무게(테스트 정답용)
```

```python
# 사이킷런에서 제공하는 선형회귀 클래스
from sklearn.linear_model import LinearRegression

lr = LinearRegression() # 선형회귀용 객체 생성하여 변수에 연결
lr.fit(train_input, train_target) # 선형회귀 훈련 시작
```

```python
print(lr.predict([[50]]))
# 인어의 무게 기울기 + 절편(y(인어의 무게) = a*x(인어의 길이)+b)
# 직선 방정식 무게 1241.83 으로 예측
```

image.png

```python
print(lr.coef_, lr.intercept_) # lr.coef_ 계수, lr.intercept_ 인터셉트
# 기울기를 종종 계수(coefficent), 가중치(weight)라고도 부른다.
```

```python
import matplotlib.pyplot as plt

# 훈련 세트의 산점도를 그립ㄴ비다.
plt.scatter(train_input, train_target)
# 15에서 50까지 1차 방정식 그래프를 그립니다
plt.plot([15,50], [15*lr.coef_+lr.intercept_, 50*lr.coef_+lr.intercept_])
# 직선차트 x축값                 y축값[15*계수+인터셉트, 50*계수+인터셉트]
#                인어의 무게공식 = 기울기*인어의길이+절편

# 50cm 인어 데이터

plt.scatter(50, 1241.8 , marker='^')
plt.xlabel('길이')
plt.ylabel('무게')
plt.show()
```

```python
# 결론 예측은 가능한데 직선이라 값이 너무 직선적이다.

print(lr.score(train_input, train_target))  # 93.98%
print(lr.score(test_input, test_target))    # 82.47%
# 훈련세트가 테스트에 비해 높지만 차이가 많이 남으로 과소적합이라고 함 (5%내외가 좋다.)
```

```python
# 다항회귀(Polynomial Regression) 제곱을 활용해서 직선을 곡선으로 처리가능
```

image.png

```python
# 다항회귀는 2차원 방정식의 그래프로 그리려면 길이를 제곱한 항이 훈련세트에 추가되어야 함

# 선형회귀의 문제점을 왼쪽하단에 있는 기준점이 불일치 하다 -> 곡선으로 해야함

train_poly = np.column_stack((train_input **2, train_input))
# 제곱하여 두 배열을 나란히 붙인다
test_poly = np.column_stack((test_input **2, test_input))
print(train_poly.shape, test_poly.shape)
# 2개의 특성이 만들어짐 (제곱, 원본)
```

```python
lr = LinearRegression() #선형회귀 객체생성
lr.fit(train_poly, train_target) # 훈련시작
print(lr.predict([[50**2, 50]])) # 결과 테스트

# 테스트 할 때 이 모델의 길이의 제곱과 원래 길이를 함께 넣음
# 결과 50cm의 무게는 1573.98g으로 나옴
# 직선형에서는 1241.83g
```

```python
# 인어의 무게 기울기 + 절편
# (y(인어의 무게) = a * x(인어의 길이) +b)
print(lr.coef_, lr.intercept_) # lr.coef_ 계수, lr.intercept_ 인터셉트
# 1573.98 = 1.01         * 50**2            - 21.55 * 50 + 116.05
# 무게 = lr.coef_[0] * 농어의길이 **2 - lr.coef_[1] * 50 + lr.intercept_
```

```python
# 구간별 직선을 기리기 위해 15에서 49까지 정수배열을 만든다
point = np.arange(15,50)
# 훈련세트의 산점도 표시
plt.scatter(train_input, train_target)
# 2차 방정식 그래프 표시
plt.plot(point, 1.01*point**2 - 21.6*point + 116.05)
# 50cm 인어
plt.scatter([50], [1574], marker= 'D')
plt.xlabel('길이')
plt.ylabel('무게')
plt.show()
```
